```python
def outlier_limit(col):
    Q3,Q1=np.nanpercentile(col,[75,25])
    IQR=Q3-Q1
    UL=Q3+1.5*IQR
    LL=Q1-1.5*Q1
    return Ul,LL
    
```

```python
for column in data_cle.columns:
    if data_cle[column].dtype !='object':
        if i not in list1:
            if pd.api.types.is_numeric_dtype(data_cle[i]):
                UL,LL=outlier_limit(data_cle[column])
                data_cle[column]=np.where((data_cle[column]>UL)| (data_cle[column]<LL),np.nan,data_cle[i])
```
