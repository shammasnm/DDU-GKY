# Adding missing data to individual columns

#### The same method can be uesed to add missing data for various columns differently.We just need to pass a dictionary as below


```python
wind.fillna({
    "Temperature":0,
    "Windspeed":5,
    "Status":"sunny"
})
```

# Forward fill (row)

```python
wind.fillna(method="ffill")
```

# Backward fill(row)

```python
wind.fillna(method="bfill")
```

# Limiting the forward/backward fill

```python
wind.fillna(method="ffill",limit=1)
```

# Filling with pandas objects

```python
wind.fillna(wind.mean())
```

# Filling for specific range of columns

```python
wind.fillna(wind.mean()["Temperature":"Windspeed"])
```

# interpolate missing value

```python
wind.interpolate()
```

# Time interpolate

```python
wind.interpolate(limit=1,limit_direction="backward")
```

# limit area of interpolation

```python
wind.interpolate(limit_direction="both",limit_area="inside",limit=1)
```
