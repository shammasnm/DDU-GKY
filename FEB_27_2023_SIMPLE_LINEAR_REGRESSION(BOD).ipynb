```python
import pandas as pd 
import numpy as np 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
plt.style.use("dark_background")
```

```python
from pydataset import data
```

```python
data()
```

```python
df=data("longley")
df
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
x=df[["time"]].values
x
```

```python
data()
```

```python
y=df[["BJsales"]].values
```

```python
sc=StandardScaler()
```

```python
x=sc.fit_transform(x)
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
traing_score=model.score(x_train,y_train)*100
traing_score
```

```python
R2_score=r2_score(y_test,model.predict(x_test))*100
R2_score
```

```python
mae=mean_absolute_error(y_test,model.predict(x_test))
mae
```

```python
mse=mean_squared_error(y_test,model.predict(x_test))
mse
```

```python
y_pres=model.predict(x)
```

```python
y_pres
```

```python
output=pd.DataFrame(zip(y,y_pres),columns=("actual","predict"),dtype=float)
output
```

```python
plt.figure(figsize=(12,6))
plt.title("PREDICTED VS ACTUAL DATA")
plt.xticks()
sns.scatterplot(data=output,x="actual",y="predict");
```
